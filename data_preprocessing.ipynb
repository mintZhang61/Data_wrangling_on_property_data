```python
# make the consistence of salary type
s = pd.Series(ufo['Salary per annum'])
# change all types of "xxK" to "xx000"
par1 = re.compile('\d{2}K')
for i in range(len(s)):
    if re.findall(par1, s[i]):
        s[i] = s[i][:-1]
        s[i]= s[i] + "000"
    
dfff = pd.DataFrame(s)
#dfff.ix[4070:4100, 0]
ufo['Salary per annum'] = ufo['Salary per annum'].str.extract('(\d*)', expand = True)
ufo.ix[3920:4000, 7]
```

### f. Location

```python
#check any Lexical errors in Location by similarity checking
def check_similarity(a,b):
    return SequenceMatcher(None, a,b).ratio()
location = list(ufo['Location'].unique())
location.sort()
for index, i in enumerate(location):
    for j in location[index+1:]:
        if check_similarity(i,j) > 0.90:
            print(i+":"+j)
```

After checking those places, the last three are lexical errors

```python
# Lexical errors
ufo.Location.replace(['Nottinham','Oxfords','Reeding','Surey'],
                     ['Nottingham','Oxford','Reading','Surrey'], inplace = True)
                      
                          
```
